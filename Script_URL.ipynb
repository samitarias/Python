```python
import requests
from datetime import datetime
import time

# Lista de URLs a las que enviarás las peticiones
urls = [
    'https://translatorapi.davivienda.cr/translatorApi/codeTranslation?name1=IdentificationType&&esp=TECH&sp=NSBT&unpaged=true',
    'https://translatorapi.davivienda.cr/translatorApi/codeTranslation?name1=IdentificationType&&esp=TECH&sp=NSBT&unpaged=true',
    'https://translatorapi.davivienda.cr/translatorApi/tmf/enumSet/AgrupationProduct/PRODUCTOS_PAGO_TERCEROS?channel=IN&locale=es_CR'
    # Agrega aquí las URLs que necesites
]

# Función para enviar la solicitud y obtener la información
def obtener_info(url):
    try:
        # Realiza la solicitud HTTP GET
        response = requests.get(url)

        # Obtén el código de estado (código de respuesta HTTP)
        codigo = response.status_code

        # Obtén la fecha y hora de la solicitud
        fecha_hora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Imprime los resultados
        print(f"URL: {url}")
        print(f"Código de Respuesta: {codigo}")
        print(f"Fecha y Hora: {fecha_hora}")
        print("-" * 40)
    except requests.exceptions.RequestException as e:
        print(f"Error al intentar acceder a {url}: {e}")
        print("-" * 40)

# Bucle infinito que ejecuta las solicitudes indefinidamente
while True:
    for url in urls:
        obtener_info(url)

    # Espera un tiempo antes de la siguiente iteración
    time.sleep(10)  # 10 segundos de espera antes de la siguiente ronda
```
